TASK 19: import the "insurance.csv" file and plot a scatter plot and a regression line that predicts how age affects costs.

```python
# Import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Import and display the dataset
df = pd.read_csv("insurance.csv")
df[:]
```

```python
# Using the iloc function, assign column "age" to the x-axis and column "Charges" to the y-axis.
x = df.iloc[:,:1].values 
y = df.iloc[:,-1].values
# set the colour of the dots to 'blue' and name the axes
plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# name the model and fit it to the data using "linear_model.LinearRegression() from sklearn
charges_model = LinearRegression()
charges_model.fit(x,y)
```

```python
# Plot the data on a scatter plot with the best-fit regression line
y_pred = charges_model.predict(x)
# Set the colour of the dots to 'blue' and the colour of the line to red.
plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
# Name the axes
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Let's try to predict an unknown value by adding a random age value to the dataset with the append and reshape methods
unk_x = [[38]] 
x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = charges_model.predict(x_pred)

print("The cost for a 38-year old customer should be:", charges_model.predict(unk_x))
```
